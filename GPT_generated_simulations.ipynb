```python
'''The following is the simulation of the store's inventory system written in Python.
The simulation starts at day 0. The initial storage level is 50. The replenishment frequency is 9 days. The replenishment size is 110.
Demand is a random variable under the uniform distribution, taking values in range from 10 to 16. The simulation lasts 60 days.
Demand and inventory levels are plotted after the simulation. The grid and legend must be on that plot.
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Initialization
storage_level = 50
replenishment_frequency = 9
replenishment_size = 110
demand_min = 10
demand_max = 16
#Simulation
storage_level_list = [storage_level]
replenishment_list = [0]
demand_list = [0]
for i in range(60):
  demand = np.random.uniform(demand_min, demand_max)
  demand_list.append(demand)
  if i % replenishment_frequency == 0:
    storage_level = storage_level + replenishment_size
    replenishment_list.append(i)
  storage_level = storage_level - demand
  storage_level_list.append(storage_level)
#Plotting
plt.plot(storage_level_list, label = 'Storage level')
plt.plot(replenishment_list, [50]*len(replenishment_list),
         'o', label='Replenishment')
plt.plot(demand_list, label='Demand')
plt.grid()
plt.legend()
plt.show()


```

```python
'''The following is a simulation of an inventory management system of the store written in Python.
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
storage = 50
time = 0
end_time = 90
inventory_levels = []
replenishment_frequency = 9
demand = np.random.uniform(10, 16)
replenishment_size = 50
xlabel = 'time'
ylabel = 'inventory'


while time < end_time:
  inventory_levels.append(storage)
  time += 1
  if time % replenishment_frequency == 0:
    storage += replenishment_size
  demand = np.random.uniform(10, 16)
  storage -= demand
  if storage < 0:
    storage = 0

plt.plot(inventory_levels)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
time = 0.0
queue = []
queue.append('customer') #1st customer
queue.append('customer') #2nd customer
queue.append('customer') #3rd customer
timings = [time]
queues = [len(queue)]


def interarrival():
  return random.exponential(2.1)
def service():
  return random.exponential(1.9)
arr = interarrival()
dep = service()
while time < 200:
  if arr < dep:
    time += arr
    arr += interarrival()
    queue.append('customer')
  else:
    time += dep
    dep += service()
    queue.pop(0)
  timings.append(time)
  queues.append(len(queue))
plt.plot(timings, queues)
plt.xlabel('Time')
plt.ylabel('Queue length')
plt.grid(True)
plt.legend(['Queue length'])
plt.show()
```
