```python
import numpy as np
from numpy import random 
import matplotlib.pyplot as plt
```

```python
class SKU():
  def __init__(self, a=7, b=10, c=15, lead=7, target=80):
    self.time = 0 #meassured in days
    self.a = a 
    self.b = b 
    self.c = c 
    self.lead = lead 
    self.target = target
    self.Q = target
    self.I = 80
    self.history = [[], [], []] 
  def generate_demand(self):
    return random.triangular(self.a, self.b, self.c)
  def simulate(self):
    self.time += 1
    daily_demand = self.generate_demand()
    if self.time % 7 == 0:
      tmp = self.I
      self.I += self.Q
      self.Q = self.target - tmp  
    if daily_demand < self.I:
      self.I -= daily_demand
    else:
      self.I = 0
    self.history[0].append(self.time)
    self.history[1].append(self.I)
    self.history[2].append(daily_demand)
  def plotter(self):
    plt.plot(self.history[0], self.history[1])
    plt.grid(True)
    plt.xlabel('Days')
    plt.ylabel('Inventory')
```

```python
sku1 = SKU()
while sku1.time <= 30:
  sku1.simulate()
```

```python
sku1.plotter()
```
